```python
import requests
import wikipediaapi

# Fonction pour effectuer une recherche sur Wikipédia
def search_wikipedia(query, lang="fr"):
    URL = f"https://{lang}.wikipedia.org/w/api.php"
    params = {
        "action": "opensearch",
        "search": query,
        "limit": 1,         # On prend le 1er résultat
        "namespace": 0,
        "format": "json"
    }
    headers = {
        "User-Agent": "MonScript/1.0 (https://votre-site.com/contact)"
    }
    response = requests.get(URL, params=params, headers=headers)
    data = response.json()
    if len(data) >= 2 and data[1]:
        return data[1][0]  # Le premier titre trouvé
    return None

# Initialisation de l'instance Wikipedia pour le français
wiki_fr = wikipediaapi.Wikipedia(
    language='fr',
    user_agent='MonScript/1.0 (https://votre-site.com/contact)'
)

# Supposons que df est votre DataFrame et que la colonne contenant le nom du film est "nom film"
found = 0
not_found = 0

for index, row in df.iterrows():
    titre_fr = row['nom film']

    # Construire la requête en simulant la saisie manuelle
    # Vous pouvez essayer différentes concaténations, par exemple "NomDuFilm film" ou "NomDuFilm (film)"
    requete = f"{titre_fr} film"
    titre_recherche = search_wikipedia(requete, lang="fr")

    if titre_recherche:
        page_fr = wiki_fr.page(titre_recherche)
        if page_fr.exists():
            # Traitement si la page a été trouvée
            found += 1
            # Optionnel : enregistrer le titre trouvé dans le DataFrame
            df.at[index, 'Titre_wiki'] = page_fr.title
            print(f"Film trouvé : {page_fr.title}")
        else:
            not_found += 1
            print(f"La page pour '{titre_fr}' (recherche: '{titre_recherche}') n'existe pas.")
    else:
        not_found += 1
        print(f"Aucun résultat trouvé pour la recherche: '{requete}'.")

print("Nombre de pages trouvées :", found)
print("Nombre de pages non trouvées :", not_found)
```

Stratégie:
On parcourt tous les films du df des meilleures audiences
Pour chaque film:
- Si on le retrouve dans les données externes, on récupère cette ligne
- Sinon on le traduit en anglais et on récupère cette ligne


```python
import requests
import wikipediaapi

# Fonction pour effectuer une recherche sur Wikipédia
def search_wikipedia(query, lang="fr"):
    URL = f"https://{lang}.wikipedia.org/w/api.php"
    params = {
        "action": "opensearch",
        "search": query,
        "limit": 1,         # On prend le 1er résultat
        "namespace": 0,
        "format": "json"
    }
    headers = {
        "User-Agent": "MonScript/1.0 (https://votre-site.com/contact)"
    }
    response = requests.get(URL, params=params, headers=headers)
    data = response.json()
    if len(data) >= 2 and data[1]:
        return data[1][0]
    return None


wiki_fr = wikipediaapi.Wikipedia(
    language='fr',
    user_agent='MonScript/1.0 (https://votre-site.com/contact)'
)

df_final = pd.DataFrame(columns=df2.columns)

found = 0
not_found = 0

for index, row in df.iterrows():
    print(f'\r{index}/{len(df)}\t\t',end='')
    titre_fr = row['nom film']

    matching_rows = df2[df2['Title'] == titre_fr]
    if not matching_rows.empty:
        df_final = pd.concat([df_final, matching_rows.iloc[[0]]], ignore_index=True)
    else:
        requete = f"{titre_fr}"
        titre_recherche = search_wikipedia(requete, lang="fr")

        if titre_recherche:
            page_fr = wiki_fr.page(titre_recherche)
            if page_fr.exists():
              if 'en' in page_fr.langlinks:
                  page_en = page_fr.langlinks['en']
                  titre_en = page_en.title
                  matching_rows = df2[df2['Title'] == titre_en]
                  if not matching_rows.empty:
                      df_final = pd.concat([df_final, matching_rows.iloc[[0]]], ignore_index=True)
                      found += 1
            else:
                not_found += 1
        else:
            not_found += 1

print("Nombre de pages trouvées :", found)
print("Nombre de pages non trouvées :", not_found)

```
